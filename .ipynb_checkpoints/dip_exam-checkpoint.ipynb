1. Find the Beluga Whale contours in the  images.(5%)
2. Calculate the countours area.(5%)

```python
img_a = cv2.imread("img/7.jpg")
img_a_copy = cv2.imread("img/7.jpg")

img_a = cv2.cvtColor(img_a, cv2.COLOR_BGR2RGB)
img_a_copy = cv2.cvtColor(img_a_copy, cv2.COLOR_BGR2RGB)

imgbw = cv2.cvtColor(img_a, cv2.COLOR_BGR2GRAY)
_, img_thresh = cv2.threshold(imgbw, 30, 255, cv2.THRESH_BINARY)

kernel = np.ones((11, 11), np.uint8)
img_clean = cv2.morphologyEx(img_thresh, cv2.MORPH_OPEN, kernel)

contours, _ = cv2.findContours(img_clean, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create a blank image to draw contours
result = np.zeros_like(img_thresh)

# Initialize areas
area = 0

# Loop through contours and draw them
for contour in contours:
    area = cv2.contourArea(contour)
    print(area)
    if area > 1000:  # Threshold to distinguish pad from probe marks (adjust as needed)
        area += area
        cv2.drawContours(result, [contour], -1, (255), thickness=cv2.FILLED)  
        cv2.drawContours(img_a_copy, [contour], -1, (0, 255, 0), thickness=10)


img_b = img_a_copy
countours_area = area

# show image
plt.imshow(img_a)
plt.show()
plt.imshow(img_b, 'gray')
plt.show()
print('The Beluga Whale area is ',countours_area)
```
